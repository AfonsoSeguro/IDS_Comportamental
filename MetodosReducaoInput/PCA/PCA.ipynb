## STEP 1: "STANDARDIZATION"

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaler.fit(input_label)
input_label = scaler.transform(input_label)
```

## STEP 2: COVARIANCE MATRIX COMPUTATION

```python
def media(X):
    t = 0
    for x in X:t += x
    return t / len(X)
```

```python
def calculoCovariancia(x, y):
    n = len(x);
    media_x = media(x)
    media_y = media(y)
    t = 0
    for i in range(n):
        t += ((x[i] - media_x)*(y[i] - media_y))
    return t / n
```

```python
n = len(input_label[0])
cov_matriz = [] 
for x in range(n):
    cov_matriz.append([])
    for y in range(n):
        cov_matriz[x].append(calculoCovariancia(input_label[x], input_label[y]))
```
